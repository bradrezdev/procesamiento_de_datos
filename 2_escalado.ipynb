# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

### 1. ¿Qué diferencias encuentras entre las estadísticas originales y las del escalado MinMax?


```python
# Comparar estadísticas originales vs MinMax
print('=== Estadísticas originales ===')
print(data.describe())
print()
print('=== Estadísticas con MinMaxScaler ===')
print(data_minmax.describe())
print()
print('Diferencias principales:')
print('- Original: cada variable tiene su propio rango (min y max diferentes)')
print('- MinMax: todas las variables quedan en el rango [0, 1]')
print('- Original: la media y desviación estándar dependen de cada variable')
print('- MinMax: min=0, max=1, pero media y std varían según la distribución')
```

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
# Gráfica de distribución: sepal length antes y después del escalado
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

# Antes del escalado
plt.subplot(1, 3, 1)
plt.hist(data['sepal length (cm)'], bins=15, color='steelblue', edgecolor='black')
plt.title('Original')
plt.xlabel('sepal length (cm)')
plt.ylabel('Frecuencia')

# Después de MinMaxScaler
plt.subplot(1, 3, 2)
plt.hist(data_minmax['sepal length (cm)'], bins=15, color='orange', edgecolor='black')
plt.title('MinMaxScaler')
plt.xlabel('sepal length (escalado)')

# Después de StandardScaler
plt.subplot(1, 3, 3)
plt.hist(data_standard['sepal length (cm)'], bins=15, color='green', edgecolor='black')
plt.title('StandardScaler')
plt.xlabel('sepal length (escalado)')

plt.tight_layout()
plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
# Gráfica de dispersión: sepal length vs sepal width con StandardScaler
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

# Antes del escalado
plt.subplot(1, 2, 1)
plt.scatter(data['sepal length (cm)'], data['sepal width (cm)'], alpha=0.6)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Original')

# Después de StandardScaler
plt.subplot(1, 2, 2)
plt.scatter(data_standard['sepal length (cm)'], data_standard['sepal width (cm)'], alpha=0.6, color='green')
plt.xlabel('sepal length (escalado)')
plt.ylabel('sepal width (escalado)')
plt.title('StandardScaler')

plt.tight_layout()
plt.show()

print('Nota: La forma de la dispersión se mantiene igual,')
print('pero los valores cambian de escala. Con StandardScaler,')
print('ambas variables quedan centradas en 0 con desviación estándar 1.')
```
